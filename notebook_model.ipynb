```python
from imblearn.over_sampling import RandomOverSampler

# Pisahkan fitur (X) dan target (y)
X = data_cleaned.drop('Status', axis=1)  # Ganti 'Status' dengan nama kolom target
y = data_cleaned['Status']              # Ganti 'Status' dengan nama kolom target

# Terapkan Random Over-Sampling
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

# Buat DataFrame baru dari data yang sudah seimbang
data_balanced = pd.DataFrame(X_resampled, columns=X.columns)
data_balanced['Status'] = y_resampled # Ganti 'Status' dengan nama kolom target Anda

print("\nDistribusi nilai dalam kolom 'Status' setelah penyeimbangan:")
print(data_balanced['Status'].value_counts())

print("\nUkuran data setelah penyeimbangan:")
data_balanced.shape
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Identifikasi kolom kategorikal dan numerik
categorical_cols = data_balanced.select_dtypes(include=['object']).columns
numerical_cols = data_balanced.select_dtypes(include=['number']).columns

# 1. Encoding Fitur Kategorikal
# Gunakan Label Encoding untuk kolom kategorikal
label_encoders = {}
for col in categorical_cols:
    # Hindari encoding pada kolom target jika masih ada di sini setelah oversampling
    if col != 'Status': # Ganti 'Status' dengan nama kolom target Anda
        le = LabelEncoder()
        data_balanced[col] = le.fit_transform(data_balanced[col])
        label_encoders[col] = le # Simpan encoder jika perlu inverse_transform nanti

print("\nData setelah Label Encoding fitur kategorikal:")
print(data_balanced.head())

# 2. Normalisasi Fitur Numerik
# Gunakan StandardScaler untuk kolom numerik
scaler = StandardScaler()
data_balanced[numerical_cols] = scaler.fit_transform(data_balanced[numerical_cols])

print("\nData setelah Normalisasi fitur numerik:")
print(data_balanced.head())
```

```python
# Pisahkan fitur (X) dan label (y)
X = data_balanced.drop('Status', axis=1)  # Semua kolom kecuali 'Status' adalah fitur
y = data_balanced['Status']              # Kolom 'Status' adalah label

# Ubah label 'Status' menjadi biner: Alive = 1, Dead = 0
# Pastikan nilai di kolom 'Status' sudah dalam format yang konsisten
# Berdasarkan output sebelumnya, nilai 'Status' sudah berupa 'Alive' dan 'Dead'
y = y.map({'Alive': 1, 'Dead': 0})

print("\nFitur (X):")
print(X.head())

print("\nLabel (y):")
print(y.head())

print("\nDistribusi label biner (y):")
print(y.value_counts())
```

```python
from sklearn.model_selection import train_test_split

# Split data menjadi training dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

print("\nUkuran data training (Fitur):", X_train.shape)
print("Ukuran data testing (Fitur):", X_test.shape)
print("Ukuran data training (Label):", y_train.shape)
print("Ukuran data testing (Label):", y_test.shape)

print("\nDistribusi label di training set:")
print(y_train.value_counts())

print("\nDistribusi label di testing set:")
print(y_test.value_counts())
```

```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Tentukan daftar nilai k yang akan diuji
k_values = range(1, 22, 2) # Bilangan ganjil dari 1 hingga 21

# Inisialisasi list untuk menyimpan akurasi testing
accuracy_scores = []

# Loop melalui setiap nilai k
for k in k_values:
    # Inisialisasi model K-NN dengan nilai k saat ini
    knn_model = KNeighborsClassifier(n_neighbors=k)

    # Latih model menggunakan data training
    knn_model.fit(X_train, y_train)

    # Lakukan prediksi pada data testing
    y_pred = knn_model.predict(X_test)

    # Hitung akurasi pada data testing
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

    print(f"Akurasi dengan k={k}: {accuracy:.4f}")

# Temukan nilai k dengan akurasi tertinggi
best_k_index = accuracy_scores.index(max(accuracy_scores))
best_k = k_values[best_k_index]
best_accuracy = accuracy_scores[best_k_index]

print(f"\nNilai k terbaik berdasarkan akurasi pada data testing adalah {best_k} dengan akurasi {best_accuracy:.4f}")

# (Opsional) Visualisasikan akurasi untuk setiap nilai k
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores, marker='o', linestyle='-')
plt.title('Akurasi K-NN pada Data Testing vs. Nilai k')
plt.xlabel('Nilai k')
plt.ylabel('Akurasi Testing')
plt.xticks(k_values)
plt.grid(True)
plt.show()
```

```python
import numpy as np

def euclidean_distance(row1, row2):
  """Hitung jarak Euclidean antara dua baris data."""
  distance = 0.0
  for i in range(len(row1) - 1): # Tidak termasuk kolom target
    distance += (row1[i] - row2[i])**2
  return np.sqrt(distance)

def get_neighbors(train, test_row, num_neighbors):
  """Ambil k tetangga terdekat dari data training untuk data uji."""
  distances = []
  for train_row in train:
    dist = euclidean_distance(test_row, train_row)
    distances.append((train_row, dist))
  distances.sort(key=lambda tup: tup[1]) # Urutkan berdasarkan jarak
  neighbors = [neighbor[0] for neighbor in distances[:num_neighbors]] # Ambil k tetangga
  return neighbors

def predict_classification(train, test_row, num_neighbors):
  """Tentukan kelas prediksi berdasarkan mayoritas tetangga."""
  neighbors = get_neighbors(train, test_row, num_neighbors)
  output_values = [row[-1] for row in neighbors] # Ambil label dari tetangga
  prediction = max(set(output_values), key=output_values.count) # Tentukan kelas mayoritas
  return prediction

# Konversi DataFrame ke numpy array untuk memudahkan perhitungan jarak
X_train_np = X_train.values
X_test_np = X_test.values
y_train_np = y_train.values

# Gabungkan fitur dan label untuk data training agar sesuai format fungsi get_neighbors
train_dataset = np.hstack((X_train_np, y_train_np.reshape(-1, 1)))

# Contoh penggunaan dengan satu data uji (baris pertama dari X_test)
test_instance = X_test_np[0]
num_neighbors_example = best_k # Gunakan nilai k terbaik yang sudah ditemukan

# Hitung jarak Euclidean dari data uji ke seluruh data latih dan ambil k data terdekat
neighbors_example = get_neighbors(train_dataset, test_instance, num_neighbors_example)

print(f"\nTetangga terdekat untuk data uji pertama dengan k={num_neighbors_example}:")
for neighbor in neighbors_example:
    print(f"  Data: {neighbor[:-1]}, Label: {neighbor[-1]}, Jarak: {euclidean_distance(test_instance, neighbor)}")

# Tentukan kelas berdasarkan mayoritas tetangga
prediction_example = predict_classification(train_dataset, test_instance, num_neighbors_example)
print(f"\nPrediksi kelas untuk data uji pertama: {prediction_example}")
print(f"Label sebenarnya: {y_test.iloc[0]}")
```

```python
from sklearn.metrics import confusion_matrix

# Lakukan prediksi pada data testing menggunakan model terbaik (dengan best_k)
knn_best_model = KNeighborsClassifier(n_neighbors=best_k)
knn_best_model.fit(X_train, y_train)
y_pred_best = knn_best_model.predict(X_test)

# Hitung confusion matrix
cm = confusion_matrix(y_test, y_pred_best)

# Cetak confusion matrix
print("\nConfusion Matrix:")
print(cm)

# Ekstrak nilai dari confusion matrix
# cm layout:
# [[TN, FP],
#  [FN, TP]]
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]
TP = cm[1, 1]

# Hitung akurasi menggunakan rumus
accuracy_manual = (TP + TN) / (TP + TN + FP + FN)

print(f"\nTP: {TP}")
print(f"TN: {TN}")
print(f"FP: {FP}")
print(f"FN: {FN}")

print(f"\nAkurasi (dihitung manual): {accuracy_manual:.4f}")

# Verifikasi dengan accuracy_score dari sklearn (seharusnya sama dengan best_accuracy)
accuracy_sklearn = accuracy_score(y_test, y_pred_best)
print(f"Akurasi (menggunakan sklearn): {accuracy_sklearn:.4f}")
```
